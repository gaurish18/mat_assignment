Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

ans-Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in a wide variety of formats and quality levels. It provides a flexible and comprehensive set of tools for creating various types of plots, charts, and graphs.

Matplotlib can be used for a range of tasks, including data exploration, data analysis, scientific visualization, and even creating publication-quality graphics. It is widely used in fields such as data science, engineering, physics, finance, and more.

The Pyplot module within Matplotlib provides a simple interface for creating and customizing plots. It offers a set of functions that allow you to create various types of plots. Here are five types of plots that can be created using the Pyplot module of Matplotlib:

1.Line Plot: Line plots are used to visualize the relationship between two continuous variables. They are suitable for showing trends and patterns over time or any other continuous dimension.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

2.Scatter Plot: Scatter plots are used to display individual data points as dots. They are useful for identifying relationships and patterns between two variables.

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

3.Bar Plot: Bar plots are used to compare categorical data by representing each category as a rectangular bar. They are great for showing the distribution of data among different categories.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

Q2.What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Q3.Why is the subplot() function used? Draw four line plots using the subplot() function.

ans-The subplot() function in Python is used to create a grid of subplots within a single figure. This allows you to display multiple plots or charts within the same figure, making it easier to compare and analyze data. subplot() takes three arguments: the number of rows in the grid, the number of columns in the grid, and the index of the current subplot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the data for each line plot
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1 data
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2 data
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3 data
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4 data
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# Subplot 1: Top left
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Subplot 2: Top right
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Subplot 3: Bottom left
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Subplot 4: Bottom right
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('company_vs_profit')
plt.show() 
```

```python
#Horizontal _bar_chart
plt.barh(company, profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('company_vs_profit')
plt.show() 
```

Q5.What is a box plot? Why is it used? Using the following data plot a box plot.

ans-A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and potential outliers of the data. Box plots are commonly used for the following purposes:

1.Visualizing Data Distribution: Box plots provide a concise way to visualize the spread and skewness of a dataset, making     it easy to identify central tendencies and variations.

2.Comparing Distributions: Box plots are useful for comparing the distributions of multiple datasets, allowing you to         quickly assess differences in medians, quartiles, and the spread of data.

3.Detecting Outliers: Outliers can be identified as individual points beyond the "whiskers" of the plot, making it easier to   spot data points that may be significantly different from the rest of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.xlabel('Box')
plt.ylabel('Value')
plt.show
```
